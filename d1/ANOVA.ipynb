```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
train_sample_path = './input/train.csv'
df = pd.read_csv(train_sample_path)

# Remove 'ID' column and separate features and target
X = df.drop(columns=['ID', 'SUBCLASS'])
y = df['SUBCLASS']

X.replace('WT', 0, inplace=True)
mutation_cols =X.columns
# Encode the target labels (SUBCLASS) to numeric values
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
le_mutations = LabelEncoder()
# WT는 0으로 변환하고, 나머지는 LabelEncoder로 변환
def encode_mutations(value):
    if value == 'WT':
        return 0
    else:
        return le_mutations.fit_transform([value])[0]

for col in mutation_cols:
    X[col] = X[col].apply(encode_mutations)
# Perform ANOVA F-test to select top k features
from sklearn.feature_selection import SelectKBest, f_classif
k = 50  # Selecting the top 50 features for now
anova_selector = SelectKBest(score_func=f_classif, k=k)
X_selected = anova_selector.fit_transform(X, y_encoded)

# Get the selected feature names
selected_feature_indices = anova_selector.get_support(indices=True)
selected_feature_names = X.columns[selected_feature_indices]

# Display the selected feature names
selected_feature_names
```
