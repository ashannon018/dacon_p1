```python
import pandas as pd

# 예시 데이터프레임 생성
df = pd.DataFrame({
    'col1': ['apple', 'WT', 'banana', 'wtcherry'],
    'col2': ['orange', 'WT', 'peach', 'plum'],
    'col3': ['WT', 45.6, 'wtapple', 'pear']
})

# 각 셀에 대해 조건을 적용하여 값 변경 (value가 'WT'면 '', 아니면 컬럼명+첫 글자/숫자)
def process_cell(value, col_name):
    if value == 'WT':  # 값이 'WT'이면 빈 문자열
        return ''
    elif isinstance(value, str):  # 문자열일 때는 첫 글자 + 컬럼명
        return col_name + value[0]
    elif isinstance(value, (int, float)):  # 숫자일 때는 첫 숫자(소수점 전) + 컬럼명
        return col_name + str(value)[0]
    return value  # 그 외에는 원래 값 반환

# 각 셀을 처리하는 함수 적용
for col in df.columns:
    df[col] = df[col].apply(lambda x: process_cell(x, col))

print(df)
```

```python
from sklearn.feature_selection import SelectKBest, chi2
df = pd.DataFrame({
    'a': [1, 0, 1, 0],
    'b': [0, 1, 0, 1],
    'c': [1, 1, 0, 1],
    'd': [0, 0, 1, 1],
    'e': [1, 0, 1, 0],
    'f': [0, 0, 0, 1],
    'g': [1, 1, 1, 0],
    'class': [0, 1, 2, 3],
    
})
# X는 입력 데이터 (특징), y는 레이블 (클래스)
X = df[['a', 'b', 'c', 'd', 'e', 'f']]
y = df['class']  # 'A' 또는 'B'의 클래스

# 카이제곱 검정을 사용하여 상위 K개의 중요한 특징 선택
selector = SelectKBest(chi2, k=3)  # 상위 3개의 중요한 컬럼 선택
X_new = selector.fit_transform(X, y)

print(X.columns[selector.get_support()])
```
